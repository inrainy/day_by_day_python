```python
from bs4 import BeautifulSoup
#원하는 부분만 출력하고 싶을 때  뷰티풀솝으로 가능
html = """
<html><body>
  <h1>파이썬으로 웹문서 읽기 </h1>
  <p> 페이지 분석기능 </p>
  <p> 페이지 정렬 </p>
</body></html>
"""

soup = BeautifulSoup(html, 'html.parser')
h1 = soup.html.body.h1
p1 = soup.html.body.p
p2 = p1.next_sibling.next_sibling
print(h1)
print(p1)
print(p2)
```

```python
html = """
<html><body>
  <h1 id = 'title'>파이썬으로 웹문서 읽기 </h1>
  <p id = 'body'> 페이지 분석기능 </p>
  <p id = 'pp'> 페이지 정렬 </p>
</body></html>
"""

soup = BeautifulSoup(html, 'html.parser')
title = soup.find(id='title')
body = soup.find(id='body')
pp = soup.find(id='pp')
print(title)
print(body)
print(pp)
```

```python
#for 문을 돌려 볼것임.

html = """
<html><body>
 <ul>
   <li><a href = "http://www.naver.com">naver</a></li>
   <li><a href = "http://www.daum.com">daum</a></li>
 </ul>
</body></html>
"""

soup = BeautifulSoup(html, 'html.parser')
links = soup.find_all("a")
for a in links:
  href = a.attrs['href']
  text = a.string
  print(href, text)


#요령만 익히고, 웹사이트의 구조만 익히면 이렇게 가져올 수 있음
```
