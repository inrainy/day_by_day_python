````python
import numpy as np
import matplotlib.pyplot as plt
w_val = []#기울기
cost_val = []#비용
xx_val = []


n_sample = 200
#x를 200개를 만들거임
x = np.random.randn(n_sample)
#x 의 샘플을 200개 랜덤으로
print(x)


y = 2 * x + 4 + np.random.randn(n_sample)
#y도 선모양의 200개를 만들거야. 근데 노이즈를 줄거야! np~

print(plt.scatter(x, y))
````

````python
num_epoch = 10 #반복 횟수
lr = 0.7 #학습 속도

w = np.random.uniform()
b = np.random.uniform()

for epoch in range(num_epoch):
  y_pred = w * x + b
  cost = np.abs(y_pred - y).mean() #절대값 abs
  xx = lr * ( (y_pred - y) * x).mean()
  #러닝레이트 만큼 y의 오차율을 구해서
  #기울기의 변화량을 구함!
  print(f'{epoch:2} w={w:.6f}, b={b:.6f}, cost={cost:.6f}, (y_pred-y).mean()={xx:.6f}')

  w = w - xx
  b = b - lr*(y_pred - y).mean()
  w_val.append(w)
  cost_val.append(cost)
  xx_val.append(xx)




  #결과값
  #w는2하고 가깝게 나옴.
  #cost는 초반에 변화하다가 후반에는 0.79정도로 수렴중인듯
  #(y_pred-y).mean() 기울기도 조금식 줄어듬!
````

````python
plt.plot(range(num_epoch), cost_val)
plt.show()

#비용이 감소되면서 포물선이 그려짐. 경사하강법에서 이런표현을 많이씀.

#기울기가 변화되는 것도 이렇게 확인이 가능할 것임.

#경사하강법이 이런식으로 진행이 되는 구나 하고 넘어가면 되나?
````

````python
# 배달거리와 배달시간 데이터
data = np.array([
    [100, 20], 
		[150, 24], 
		[300, 36], 
		[400, 47], 
		[130, 22], 
		[240, 32],
		[350, 47], 
		[200, 42], 
		[100, 21], 
		[110, 21], 
		[190, 30], 
		[120, 25], 
		[130, 18], 
		[270, 38], 
		[255, 28]])

#배달 거리와 배달 시간 이 x,y 1차원임,

x = data[:, 0]
y = data[:, 1]
````

````python
plt.xlim(50, 300)
plt.ylim(10, 60)
plt.plot(x,y, '+')
plt.grid()
````

````python
# 기울기 a를 최소제곱법으로 구하는 함수
def compute_a(x, y, mean_x, mean_y):
    # 분자 부분
    dc = 0
    for i in range(len(x)):
      dc += (x[i] - mean_x) * (y[i]- mean_y)
    # 분모 부분  
    divisor = 0
    for i in range(len(x)):
      divisor += (i - mean_x)**2


    a = dc / divisor
    return a


mx = np.mean(x)
my = np.mean(y)
````

#### 일반 선형회귀

예측값과 실제 값의 cost를 최소화할 수 있도록 W(가중치, 회귀계수)를 최적화하며, 규제(Regularization)를 적용하지 않은 모델

단순 선형회귀 (Simple Linear Regression)
> \\( H({ x })={ W }{ x }+b \\)

다변수 선형회귀 (Multi-variable Linear Regreesion)
> 변수가 3개 일때의 H(x) <br>
> \\( H({ x }_{ 1 },{ x }_{ 2 },{ x }_{ 3 })={ W }_{ 1 }{ x }_{ 1 }+{ W }_{ 2 }{ x }_{ 2 }+{ W }_{ 3 }{ x }_{ 3 }+b \\)

> 변수가 n개 일때의 H(x) <br>
>  \\( H({ x }_{ 1 },{ x }_{ 2 },{ x }_{ 3 },\dots ,{ x }_{ n })={ W }_{ 1 }{ x }_{ 1 }+{ W }_{ 2 }{ x }_{ 2 }+{ W }_{ 3 }{ x }_{ 3 }+\dots +{ W }_{ n }{ x }_{ n }+b \\)

다항 회귀 (Polynomial Regreesion)
> 회귀가 독립변수의 단항식이 아닌 2차, 3차 방정식과 같은 다항식으로 표현되는 것 <br>
> 차수가 높아질수록 과적합의 문제가 발생


````python
#다항회귀는 보통 곡선을 이야기함... 단항아니고 다항...
#x^2이면 포물선, x*11승이면 10번꺾는 포물선 뭐 이런건가봄

#복잡한모델에 많이꺾으면 모델이 복잡해짐. 모델 복잡도가 높아지고, 이럴땐 차수를 내려주는것이 필요함

#이럴때 사용하는 것이 다항회귀임.
````

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

n = 100
x = 6 * np.random.rand(n,1) -3
# np.random.rand(n,1) = 100개
y = 0.5 * x**2 + x + 2 + np.random.rand(n,1)
#w값이 1개는 0.5 하나는 1이 나올것임 왜냐 x 는 1 * x 이니까

plt.scatter(x, y, s=5)
#사이즈 5준거임
````

````python
from sklearn.preprocessing import PolynomialFeatures
#poly~ : 특성, 이해할필요없음 소개만 해줌. 나중에 다시 해주겠음!
#위에 그래프처럼 굴곡지게 만들어 주는 특성임.


poly_features = PolynomialFeatures(degree = 2, include_bias=False)
x_poly = poly_features.fit_transform(x)
x_poly.shape, x.shape
#곡선을 표현할 수 있는 표현의 속성이 하나 더 추가되어 w값이 2개가 된거임,
````

````python
from sklearn.linear_model import LinearRegression
#선형모델

#다항회귀도 가능하게 함

model = LinearRegression()
model.fit(x_poly, y)

model.coef_, model.intercept_

#두개가 들어올거고, coef는 2개로 들어올 것임.

#y = 0.5 * x**2 + x + 2
#결과값에 1과 가깝고 5과 가깝고, 2는 2를 조금 벗어남.


#곡선도 선형으로 표현할 수 있다. 라는 내용진행한것.
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import mglearn

x,y = mglearn.datasets.make_wave(60)
````

````python
plt.scatter(x,y)
````

````python
#선형 모델을 정식적으로 그려볼거임

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42)
#데이터를 나눠줄것임.

model = LinearRegression()
model.fit(x_train, y_train)

#학습시키고

print('가중치(계수, 기울기, w):', model.coef_)
print('편향(절편, b):', model.intercept_)

print('훈련 점수:', model.score(x_train, y_train))
print('테스트 점수:', model.score(x_test, y_test))
#기울기를 만든것
````
