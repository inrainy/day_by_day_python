```python
#수업, 코드 설명

def check_sound_status(a, b, r, x, y):
  if (x-a)**2 + (y-b)**2 >= r**2:
    return "silent"
  return "noisy"

  #return 반환값 가지고 나갈거야! print랑은 똑같이 출력되지만,
  #조금은 의미가 다른거니까~!
  #else로 안짤 수도 있는 거니까~!

a,b,r = map(int, input().split())
n = int(input())

for _ in range(n):
  x,y = map(int, input().split())
  print(check_sound_status(a,b,r,x,y))
```

```python
#약수 구하기
a = int(input())
measure = []
for i in range(1, a+1):
  if a % i == 0:
    #나눈 나머지가 0일때 i는 a의 약수임
    measure.append(i)

print(measure)
```

```python
#약수 구하기 + 응용
a = int(input())
measure = []
count = 0

for i in range(1, a+1):
  if a % i == 0:
    #나눈 나머지가 0일때 i는 a의 약수임
    measure.append(i)
    count += 1

print(f'{a} 의 약수는  {measure} 입니다.')
print(f'{a} 약수의 개수는 {count}개 입니다.')
```

```python
#최대값 최소값 + 강사님 코드

num_list = list(map(int, input().split()))

max_num = num_list[0]
min_num = num_list[0]

print(max_num)
print(min_num)
print('========================')
for i in num_list:
  if max_num < i:
    max_num = i
    #max_num이 1일 때, i 가 1이면 성립되지 않지만,
    #max_num이 1일 때, i 가 3이면 성립이 되서
    #i 가 max_num이 됨!!
    print(i)

  if min_num > i:
    #ex) min_num이 1인데, i가 1일 때, 1 > 1 이 성립이 안돼는
    #False임으로, 성립이 안돼서, min_num = i가 되지 않음!!
    #그냥 지나치고 출력 되는 것임
    min_num = i
print('========================')
print(max_num)
print(min_num)

#print(max(num_list))
#print(min(num_list))
```
