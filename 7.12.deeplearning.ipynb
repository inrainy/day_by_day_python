```python
#x변수를 만들어 볼것임.

#0과 7사이에 있는 무작위 값을 뽑아주기 위한 random.randint

#============================================질문

var_x = np.random.randint(0,7,size = 99)
var_x_outlier = [100]

var_x = np.hstack([var_x, var_x_outlier])

var_y = np.random.randint(0,40,size = 100)

print(var_x)
print('='*30)
print(var_y)
```

```python
#============================================질문


var_x_norm = (var_x - var_x.min()) / (var_x.max() - var_x.min())
var_y_norm = (var_y - var_y.min()) / (var_y.max() - var_y.min())

print('var_x_norm : ', var_x_norm[:10])
print('var_y_norm : ', var_y_norm[:10])
```

```python
#============================================질문
#그림이 다르다. norm? nrom?


plt.figure(figsize=(15,3))


plt.subplot(1,5,1)
sns.scatterplot(var_x, var_y, color = 'red')
plt.title('Before Min-Max Normalization')
plt.grid()


plt.subplot(1,5,3)
sns.scatterplot(var_x_norm, var_y_norm, color = 'orange')
plt.title('After Min-Max Normalization')
plt.grid()


var_x = np.random.randint(0,7, size = 100)
var_y = np.random.randint(0,40, size = 100)

var_x_norm = (var_x - var_x.min()) / (var_x.max() - var_x.min())
var_y_norm = (var_y - var_y.min()) / (var_y.max() - var_y.min())

plt.subplot(1,5,5)

sns.scatterplot(var_x_norm, var_y_norm, color = 'green')
plt.title('No Outlier Min-Max Normalization')
plt.grid()

plt.show()
```

```python
#======================? 미니배치 해준건가?
#학습비율 train_ratio
#독립변수 4
#종속변수 3

train_ratio = 0.8
input_cnt   = 4
output_cnt  = 3

train_idx = int(len(faults_np) * train_ratio)
print("train_idx : ", train_idx)
print("test_idx : ", len(faults) - train_idx)
```

```python
train_X, train_Y = faults_np[:train_idx, :-output_cnt], faults_np[:train_idx, -output_cnt : ]

test_X, test_Y = faults_np[train_idx : , : -output_cnt], faults_np[train_idx:, -output_cnt :]
```

```python
print(train_X)
print(train_Y)
print("===================")
print(test_X)
print(test_Y)
```

```python
#cross entropy 손실함수 중 하나 인 듯
#optimizer 아담이 유명하니 아담을 써 볼 거임

simple_layer_model = tf.keras.Sequential([
                                          tf.keras.layers.Dense(units=3,
                                                                activation = 'softmax',
                                                                input_shape = (4,))])
```

```python
simple_layer_model.summary()

#Trainable params : 

#Non-trainable params : 직접 만든 신경망이 아니라, 구글이나 페이스북에서 만든 신경망을 쌓고 싶을 때 제어가능
```

```python
#최적화를 어떻게 시킬 것인가.

simple_layer_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                           loss = 'categorical_crossentropy',
                           metrics = ['accuracy'])
```

```python
history_1 = simple_layer_model.fit(train_X, train_Y,
                                   epochs = 25,
                                   batch_size = 2)
```

```python
plt.figure(figsize=(12,4))
plt.plot(history_1.history['loss'], 'b--o', label = 'loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
plt.legend()
plt.show()
```

```python
train_ratio = 0.9
input_cnt = 27
output_cnt = 7

train_idx = int(len(faults_np)*train_ratio)
print('train_idx : ', train_idx)
print('test_idx : ', len(faults_np)-train_idx)
```

```python
#이걸 왜 해주어야 하는지 질문=======================================
#-output 과 지금 어디서 어디를 추출해 준 것인지.
#배열이 2차원이라, 이렇게 뽑아준 것 같음....


train_X, train_Y = faults_np[:train_idx, :-output_cnt], faults_np[:train_idx, -output_cnt:]
test_X, test_Y   = faults_np[train_idx: , :-output_cnt], faults_np[train_idx: , -output_cnt:]

print(train_X.shape)
print(test_X.shape)
```

#7.13일 deeplearning

```python
#tensorflow 불러오는 작업 진행할거임!

import tensorflow as tf

#순차적인 신경망을 만들 것이다.

multi_layer_model = tf.keras.Sequential ([
                                          tf.keras.layers.Dense(units = 48,
                                                               activation = 'elu',
                                                               input_shape = (27,)),
                                          tf.keras.layers.Dense(units = 24,
                                                                activation = 'elu'),
                                          tf.keras.layers.Dense(units = 12,
                                                                activation = 'elu'),
                                          tf.keras.layers.Dense(units = 7,
                                                                activation = 'softmax')
])
#activation 활성화 함수 elu
#shape 값이 어떻게 되냐 독립변수 몇개냐 27개,
#4개정도 계층 쌓아줄거야!
```

```python
multi_layer_model.summary()
```

```python
multi_layer_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
                          loss = 'categorical_crossentropy')

#loss 손실함수는 categorical_crossentropy
```

```python
history_2 = multi_layer_model.fit(train_X, train_Y, epochs = 200, batch_size = 10)

#몇세대에 걸쳐서 신경망을 구축할거냐 epochs
#한번의 배치에 몇개에 행을 걸칠거냐 batch size


#학습시킴
```

```python
multi_layer_model.evaluate(test_X, test_Y)
```

```python
#시각화

plt.figure(figsize=(20,4))
plt.plot(history_2.history['loss'], 'b--o', label = 'loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()

#loss값이 1에서 부터 쭈우우우욱 떨어지고 있음.

#예측하고 정확도를 구해보자! 지금부터 ㅎㅎ너무 어렵다 ㅎㅎ
```

```python
import csv
with open ('/content/drive/MyDrive/database/abalone.csv') as csvfile:
  csvreader = csv.reader(csvfile)
  next(csvreader)

  rows = []
  for row in csvreader:
      rows.append(row)




      #건너 띄워주는 함수 next
#next(csvreader)
#로우스라는 빈 함수에 넣어줄 거야
```

```python
df
```

```python
input_cnt, output_cnt = 10, 1
data = np.zeros([len(rows), input_cnt + output_cnt])

print(data.shape)
```

```python
#one hot vetor처리 한것임

for n, row in enumerate(rows):
  if row[0] == 'M': data[n,0] = 1
  if row[0] == 'F': data[n,1] = 1
  if row[0] == 'I': data[n,2] = 1
  data[n, 3:] = row[1:]

data[:5]
```

```python
#데이터를 섞어주고 학습데이터 비율만큼 쪼개주는 작업을 할거임

train_ratio = 0.8
#80%로 하기로 했음
test_begin_idx = int(data.shape[0] * train_ratio)
print("test_begin_idx : ", test_begin_idx)
```

```python
data.shape
```

```python
 shuffle_map = np.arange(data.shape[0])
 np.random.shuffle(shuffle_map)
```

```python
shuffle_map[:5]
```

```python
test_data = data[shuffle_map[test_begin_idx:]]
print(test_data.shape)
train_data = data[shuffle_map[:test_begin_idx]]
print(train_data.shape)
```

```python
test_data.shape[0] + train_data.shape[0]
```

```python
#독립과 종속으로 나눠주는 과정을 진행할 것임

train_X, train_Y = train_data[:, :-output_cnt], train_data[:, -output_cnt:]
#전체행 찍어주고 독립변수 뽑아줄거니까, 처음부터 독립변수 전까지.

test_X, test_Y = test_data[:, :-output_cnt], test_data[:, -output_cnt:]
```

```python
print(train_X.shape)
print(test_X.shape)

print(train_Y.shape)
print(test_Y.shape)
```

```python
#평균값을 먼저 만들고ㅓ, 표준편차도 필요하다!

train_X_mean = np.mean(train_X, axis = 0)
#print(train_X)

#열방향으로 뽑아줄거라서 axis = 0


#표준편차도 만들어 줄 거임
train_X_std = train_X.std(axis=0)

#다중 연산자 or복합연산자를 써줄 수 있음
```

```python
train_X -= train_X_mean
train_X /= train_X_std 
```

```python
print(train_X)
```

```python
#지금까지 학습ㄷ데이터만 수행시켰는데
#테스트데이터도 수행해 줄 것임

test_X -= train_X_mean
test_X /= train_X_std

#왜 테이스트 데이터를 학습데이터로 진행하냐?

#이유 : 테스트는 말 그대로 실제로 수집할 수 없는, 정확하게 평균값이나 표준편차 값을 알 수 없음
#왜냐 테스트데이터니까, 정확하게 구할 수 없음.
#그래서 일단은 학습데이터를 이용해서 평균과 표준편차를 활용해 주어야 함
```
